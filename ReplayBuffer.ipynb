```python
import torch
from ReplayBuffer import ReplayBuffer
# Example usage, show correct alignment during replay
episodes = 20
stored_episodes = 8
samples_per_episode = 40
device = 'cuda'

buffer = ReplayBuffer(stored_episodes, samples_per_episode, device)

for i in range(episodes):
    state = torch.tensor(i * episodes + 0.1, device=device)
    action = torch.tensor(i * episodes + 0.2, device=device)
    reward = torch.tensor(i * episodes + 0.3, device=device)
    target = torch.tensor(i * episodes + 0.4, device=device)
    for k in range (samples_per_episode):
        buffer.add(state, action, reward, target)
        state += 1.0
        action += 1.0
        reward += 1.0
        target +=1.0

    if (buffer.size >= 8):
        batch_size = 4
        seq_length = 10
        batch = buffer.random_sample(batch_size, seq_length)

        for b in batch:
            print(b)
```
